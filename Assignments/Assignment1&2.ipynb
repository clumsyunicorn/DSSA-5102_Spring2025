```python
"""
Name: Alyssia Marshall
Assignment: Assignment 1 & 2 - Locating & Exploring a Dataset in Python
Date: February 17, 2025
"""

# Load necessary packages
import pandas as pd       
import numpy as np       
import matplotlib.pyplot as plt 
import seaborn as sns    

# Setting a style for plots
sns.set_theme(style="whitegrid")

#  ASSIGNMENT #1

# Load the first dataset
df_salaries = pd.read_csv("ds_salaries.csv")  

# Preview the first few rows
print("First 5 rows of ds_salaries:")
display(df_salaries.head())

# General info: number of rows/columns, data types, etc.
print("\nGeneral Info for ds_salaries:")
df_salaries.info()

# Number of rows and columns
rows, cols = df_salaries.shape
print(f"\nThis dataset has {rows} rows and {cols} columns.")

# Column names
print("\nColumn Names in ds_salaries:")
print(df_salaries.columns.tolist())

# Statistical summary of numerical columns
print("\nStatistical Summary (Numerical Columns):")
display(df_salaries.describe())

# Plot for ds_salaries
# Salary distribution by experience level
plt.figure(figsize=(10,6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df_salaries, palette="viridis")
plt.title("Salary Distribution by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Salary in USD")
plt.show()


# Load the second dataset
df_internet = pd.read_csv("internet_users.csv")  
# Preview the first few rows
print("\nFirst 5 rows of internet_users:")
display(df_internet.head())

# General info: number of rows/columns, data types, etc.
print("\nGeneral Info for internet_users:")
df_internet.info()

# Number of rows and columns
rows_inet, cols_inet = df_internet.shape
print(f"\nThis dataset has {rows_inet} rows and {cols_inet} columns.")

# Column names
print("\nColumn Names in internet_users:")
print(df_internet.columns.tolist())

# Statistical summary of numerical columns
print("\nStatistical Summary (Numerical Columns):")
display(df_internet.describe())

# Quick Plot for internet_users
if 'Rate (WB)' in df_internet.columns:
    # Drop rows that don't have a valid 'Rate (WB)'
    df_plot = df_internet.dropna(subset=['Rate (WB)']).head(10)
    
    plt.figure(figsize=(10,6))
    plt.bar(df_plot['Location'], df_plot['Rate (WB)'], color='skyblue')
    plt.title("Internet Usage Rate (WB) for 10 Locations")
    plt.xlabel("Location")
    plt.ylabel("Rate (WB)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Column 'Rate (WB)' not found or is not numeric.")


#  ASSIGNMENT #2 - DATA CLEANING

# Create a copy of ds_salaries to clean
df_clean = df_salaries.copy()

# Make all column names lowercase, remove spaces & special characters
print("\nOriginal column names:")
print(df_clean.columns.tolist())

df_clean.columns = (
    df_clean.columns
    .str.lower() # make lowercase
    .str.strip() # remove leading/trailing spaces
    .str.replace(' ', '_') # replace spaces with underscores
    .str.replace(r'[^\w]', '', regex=True)  # remove any non-alphanumeric/underscore
)

print("\nCleaned column names:")
print(df_clean.columns.tolist())

# Check for missing values
missing_counts = df_clean.isnull().sum()
print("\nMissing Values in Each Column:")
print(missing_counts)

# Remove duplicates if they exist
before_dedup = len(df_clean)
df_clean.drop_duplicates(inplace=True)
after_dedup = len(df_clean)

if after_dedup < before_dedup:
    print(f"\nRemoved {before_dedup - after_dedup} duplicate rows.")
else:
    print("\nNo duplicate rows were found.")

# Save the cleaned dataset
df_clean.to_csv("ds_salaries_clean.csv", index=False)
print("\nCleaned dataset saved as ds_salaries_clean.csv")
```
